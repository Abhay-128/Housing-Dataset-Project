### <span style="color:Green"><b>Although we have the above shown null values, we cannot delete these rown as it will remove around 50% of our data, which can lead to biased or inappropriate analysis. So, we need to impute these missing values with something meaningful.</b></span>

## <span style="color:Red"><u><center><b>Missing Value Imputation.</center></u></b></span>

### <span style="color:Green">We have four columns with missing values.</span>
- Property_Name
- builder_id
- Builder_name
- Locality_Name

### <span style="color:Brown">Column to handle - 'locality_name' and 'Property_Name'</span>
#### <span style="color:Brown">We can fill null values of lacality_name with corresponding value of locality_id as both of them represents the same locality. Similarly, we can fill null values of 'Property_Name' with 'Property_id'</span>

```python
df['Property_Name'].fillna(df['Property_id'],inplace=True)
df['Locality_Name'].fillna(df['Locality_ID'],inplace=True)
df[df.columns[df.isnull().any()]].isnull().sum()
```

### <span style="color:Brown">Columns to handle - 'builder_id and builder_name'</span>

#### <span style="color:Brown">We can fill null values of 'builder_id' with 0.0 and 'Builder_name' with "Unknown".</span>

```python
df['builder_id'].fillna(0.0,inplace=True)
df['Builder_name'].fillna("Unknown",inplace=True)
```

```python
df.isnull().sum().sum()
```

## <span style="color:Red"><u><center><b>Standardisation of values - Datatype conversion</center></u></b></span>
- convert price,price_per_unit_area to integer
- rename size to size_in_sq_feet (and convert to integer)
- replace all "0 BHK" values with "< 1 BHK"

```python
df["Price"] = df["Price"].str.replace(",","")
df["Price"] = df["Price"].astype(int)
df["Price"].head()
```

```python
df["Price"].max()
```

```python
df.rename(columns={"Size":"Size_in_sq_feet"},inplace = True)
```

```python
temp = df.Size_in_sq_feet.str.split(" ",n=1,expand = True)
df["Size_in_sq_feet"] = temp[0]
df["Size_in_sq_feet"].head()
```

```python
df["Size_in_sq_feet"] = df["Size_in_sq_feet"].str.replace(",","")
df["Size_in_sq_feet"] = df["Size_in_sq_feet"].astype(int)
df.head()
```

```python
df["Price_per_unit_area"] = df["Price_per_unit_area"].str.replace(",","")
df["Price_per_unit_area"] = df["Price_per_unit_area"].astype(int)
df.head()
```

```python
df["No_of_BHK"].unique()
```

```python
df['No_of_BHK'] = df['No_of_BHK'].replace('0 BHK','< 1 BHK')
```
